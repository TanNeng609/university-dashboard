```python
import random 

def guess_number_game():
    number_to_guess = random.randint(1,100)
    input_times =0

    print ("Welcome and please guess a number between 1 and 100")

    while True:
        try:
            user_input = int(input("Please entr your guessL"))
            input_times += 1

            if user_input < number_to_guess:
                print ("Too low, try again")
            elif user_input > number_to_guess:
                print ("Too high, try again")
            else:
                print (f"Congratulations! You've guessed the number {number_to_guess} in {input_times} attempts.")
                break
        except ValueError:
            print ("Invalid input, please enter a number between 1 and 100.")


guess_number_game()
```

```python
def rock_paper_scissors():
    choices =["rock","paper","scissors"]
    print("welcome to rock paper scissors game")
    while True:
        user_choice= input("Please enter your choice (rock(0), paper(1), scissors(2)) or 'exit' to quit: ")
        if user_choice.lower() == 'exit':
            print("Thanks for playing! Goodbye!")
            break
        if user_choice not in ['0','1','2']:
            print("Invalid choice, please try again.")
            continue

        user_choice = int(user_choice)
        computer_choice = random.randint(0,2)

        print(f"You chose {choices[user_choice]}, computer chose {choices[computer_choice]}.")

        if user_choice == computer_choice:
            print("It's a tie!")

        elif (user_choice == 0 and computer_choice == 2) or \
             (user_choice == 1 and computer_choice == 0) or \
                (user_choice == 2 and computer_choice == 1):
            print("You win!")
        else:
            print("Computer wins!")

rock_paper_scissors()
```

```python
class node:
    def __init__(self,val,mark=0):
        self.val=val
        self.next=None
        self.prev=None
        self.marking=mark

class queue:
    def __init__(self):
        self.front= None
        self.rear=None
        self.size=0

    def enqueue(self,val,mark=0):
        cur_val=node(val,mark)
        self.size+=1
        if self.front==None:
            self.front=self.rear=cur_val
            self.rear.next=self.rear.prev=self.front.next=self.front.prev=self.front
        else:
            self.rear.next=cur_val
            cur_val.prev=self.rear
            self.rear=cur_val
            self.rear.next=self.front
            self.front.prev=self.rear


    def dequeue(self):
        if self.front==None:
            print("This is an empty queue")
            return -1
        
        val =self.front.val
        self.size-=1

        if self.front==self.rear:
            self.front=self.rear=None
        else:
            self.rear.next=self.front.next
            self.front.next.prev=self.rear
            self.front=self.front.next

        return val

    def deleteSpecific(self,value):
        if self.front is None:
            print("this is an empty queue")
            return -1
        
        cur = self.front
        while True:
            if cur.val==value:
                if cur==self.front==self.rear:
                    self.front=self.rear=None
                    print("all the queue is deleted")
                elif cur==self.front:
                    self.rear.next=self.front.next
                    self.front.next.prev=self.rear
                    self.front=self.front.next
                    print("the first element is deleted")
                    return
                elif cur==self.rear:
                    self.rear.prev.next=self.front
                    self.front.prev=self.rear.prev
                    self.rear=self.rear.prev
                    print("the last element is deleted")
                    return
                else:
                    cur.prev.next=cur.next
                    cur.next.prev=cur.prev
                    print("The target is deleted")
                    return
            cur = cur.next

            if cur==self.front:
                break
        print("the target element is not found in the queue")    
    def markAsDone(self,value):
        if self.front==None:
            print("This is empty queue")
            return

        cur=self.front
        while True:
            if cur.val==value:
                cur.marking=1
                print(f"Task '{value}'marked as Done!")
                return
            cur=cur.next
            if cur==self.front:
                break
        print("Task not found") 

    def printQueue(self):
        if self.front is None:
            print("\n>>> Your To-Do List is empty!<<<")
            return
        
        print("\n--- YOUR TO-DO LIST---")
        cur = self.front
        while True:
            status ="[X]"if cur.marking==1 else "[ ]"
            print(f"{status}{cur.val}")

            cur =cur.next
            if cur == self.front:
                break

        print("-------------------------------------\n")

my_list=queue()

while True:
    my_list.printQueue()
    print("1. Add a task")
    print("2. Mark Task As Done")
    print("3. Delete a specific task")
    print("4.Clear the oldest Task(dequeue)")
    print("5.Exit")

    choice = input("Choose an option (1-5): ")

    match choice:
      case 1:
            task = input("Enter new task: ")
            my_list.enqueue(task)
      case 2:
            task =input("Enter the exact name of the task you finished: ")
            my_list.markAsDone(task)
      case 3: 
            task =input("Enter the exact name of task to delete: ")
            my_list.deleteSpecific(task)
      case 4:
            removed = my_list.dequeue()
            if removed != -1:
                print(f"Removed oldest task '{removed}'")
      case 5:
            print("goodbye!!!")
            break
      case _:
            print("Invalid input, please try again.")
```

```python
import datetime
import time
import datetime
import time

def set_alarm():
    print("Python alarm clock")
    print("HH:MM:SS (24-hour clock)")

    alarm = input("Please enter your alarm time: ")
    print(f"Alarm successfully set for {alarm}. Waiting...")

    while True:
        now = datetime.datetime.now()
        current = now.strftime("%H:%M:%S")

        print(f"Current Time: {current}", end="\r")

        if current == alarm:
            print("\n\n⏰Wake UP!!!!!!!!!!⏰")
            print("The Alarm is ringing")
            print('\a\a\a')
            break
        
        # Notice how this 't' lines up perfectly with the 'i' in if, 
        # the 'p' in print, and the 'n' in now above it!
        time.sleep(1)

set_alarm()
```

```python
import json

def contact_book():
    try:
        with open("Saved_contacts.json","r") as file:
            contact=json.load(file)
    except FileNotFoundError:
        contact={}

    while True:
        print("\n--- My Contact Book ---")
        print("1. Add new contact")
        print("2.View All Contact")
        print("3.Search for a contact")
        print("4.Delete a contact")
        print("5.Exit")

        choice = input ("Choose an option(1-5)")

        match choice:
            case '1':
                name = input("Please enter your name").title().strip()
                contact_no = input("Please enter phone num: ")

                contact[name]=contact_no
                with open("Saved_contacts.json","w") as file:
                    json.dump(contact,file)


                print(f"{name} has been added to your contacts.")

            case '2':
                if len(contact) ==0:
                    print("Empty contact book")
                else:
                    print("\n---Saved Contacts ---")
                    for name,contact_no in contact.items():
                        print(f"{name}: {contact_no}")
                    print("-------------------------")
            
            case '3' :
                search_name = input("Enter the name to search for: ").title()
            
            # Checking if something exists in a dictionary is incredibly fast
                if search_name in contact:
                    print(f"Found {search_name}: {contact[search_name]}")
                else:
                    print(f"❌ {search_name} is not in your contacts.")

            case '4':
                search_name = input("Enter the name to search for: ").title()
            
            # Checking if something exists in a dictionary is incredibly fast
                if search_name in contact:
                    deleted_num = contact[search_name]
                    del contact[search_name]

                    with open("saved_contacts.json", "w") as file:
                        json.dump(contact, file)

                    print(f"Deleted {search_name}: {deleted_num}")
                else:
                    print(f"❌ {search_name} is not in your contacts.")        

            case '5':
                print("Closing")
                break

            case _:
                print("invalid choice")

contact_book()
```

```python
import turtle
import random

def draw_art():
    print("Opening canvas... watch the pen draw!")

    screen = turtle.Screen()
    screen.bgcolor("black")
    screen.title("My Python Art")

    pen =turtle.Turtle()
    pen.speed(0)
    pen.width(2)

    colors=["red","yellow","blue","green","purple","orange","cyan","magenta"]

    for i in range(150):
        pen.color(random.choice(colors))
        pen.forward(i*2)
        pen.left(144)

    pen.hideturtle()
    print("Drawing complete.")
    screen.exitonclick()

draw_art()
```

```python
import time

def show_instruction():
    print("--- 🏰 ESCAPE THE HAUNTED HOUSE 🏰 ---")
    print("Commands:")
    print("  go [direction] (north, south, east, west)")
    print("  get [item]")
    print("  quit")
    print("Goal: Find the key and reach the Garden to escape!")
    print("------------------------------------------")

def show_status(current_room,inventory,rooms):
    print("\n--------------------")
    print(f"You ar in the {current_room}")
    print(f"Inventory:{inventory}")

    if "item" in rooms[current_room]:
        print(f"You see a {rooms[current_room]['item']}")
    print("-------------------------")

def play_game():
    rooms ={
        'Hall':{
            'south':'Kitchen',
            'east':'Dining Room',
            'item':'key'
        },
        'Kitchen':{
            'north':'Hall',
            'item':'monster'
        },
        'Dining Room':{
            'west':'Hall',
            'south':'Garden',
            'item':'portion'
        },
        'Garden':{
            'north':'Dining Room'
        }
    }

    current_room='Hall'
    inventory =[]
    show_instruction()

    while True:
        show_status(current_room,inventory,rooms)

        move=input("What do you want to do? > ").lower().strip().split()

        if not move:
            continue

        action = move[0]

        if action=="quit":
            print("You gave up, thanks for playing")
            break

        elif action=="go":
            if len(move)<2:
                print("go where?(north,south,east, or west)")
                continue

            direction = move[1]
            if direction in rooms[current_room]:
                current_room=rooms[current_room][direction]
                print("Walking...")
                time.sleep(3)
            else :
                print("You cannot go that way")
        
        elif action=="get":
            if len(move)<2:
                print("Get What?")
                continue

            item = move[1]
            if 'item' in rooms[current_room] and item == rooms[current_room]['item']:
                inventory.append(item)
                print(f"You picked up the {item}!")
                del rooms[current_room]['item']
            else:
                print(f"There is no {item} there")
        
        else:
            print("I do not understand that command")

        if current_room=='Kitchen' :
            if 'portion' in inventory:
                print("\n A terrible monster hurt you and you heal yourself with the portion.")
                inventory.remove('portion')
                print("Portion used...")
                time.sleep(3)
                print("You run back to Hall")
                current_room= "Hall"
                time.sleep(3)
            else:
                print("\n💀 A terrible monster eats you! GAME OVER.")
                break

        if current_room == 'Garden':
            if 'key' in inventory:
                print("\nYou unlock the garden gate and escape!@!@@!!")
                break
            else:
                print("The gate is locked and you need to find the key")
                print("You walked back to dining room.")
                current_room='Dining Room'
                time.sleep(3)
```
